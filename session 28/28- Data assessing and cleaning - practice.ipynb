```python
patients_df = patients.copy()
treatments_df = treatments.copy()
treatments_cut_df = treatments_cut.copy()
adverse_reactions_df = adverse_reactions.copy()
```

# Define

- replace all missing values of patients_df with no data
- subtract hba1c_start to hba1c_end to get all the change values
- in patients table we will use regex to separate phone and email 

```python
patients_df.fillna('No data',inplace=True)
```

```python
patients_df.info()
```

```python
treatments_df.head()
```

```python
treatments_df['hba1c_change'] = treatments_df.hba1c_start - treatments_df.hba1c_end
```

```python
treatments_cut_df['hba1c_change'] = treatments_cut_df['hba1c_start'] - treatments_cut_df['hba1c_end']
```

```python
treatments_df.info()
```

```python
patients_df.head()
```

```python
treatments_df = pd.concat([treatments_df,treatments_cut_df])
```

```python
treatments_df = treatments_df.melt(id_vars=['given_name','surname','hba1c_start','hba1c_end','hba1c_change'],var_name='type',value_name='doasge_range')
```

```python
treatments_df = treatments_df[treatments_df['doasge_range'] != '-']
```

```python
treat = treatments_df.copy()
```

```python
treat['dosage_start'] =  treat.doasge_range.str.split('-').str.get(0)
```

```python
treat['dosage_end'] =  treat.doasge_range.str.split('-').str.get(1)
```

```python
treat.drop(columns='doasge_range',inplace=True)
```

```python
treat.dosage_start = treat.dosage_start.str.replace('u',"")
```

```python
treat.dosage_end = treat.dosage_end.str.replace('u',"")
```

```python
treatments_df = treat.merge(adverse_reactions_df,how='left',on=['given_name','surname'])
```
