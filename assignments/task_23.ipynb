### **`Problem-1:`** Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China. 

```python
# code here
df = pd.read_csv('https://tinyurl.com/2fe6vz4u')
```

```python
df.columns
```

```python
iran = df[df.Country == 'Iran'].groupby(['Country','Year'])['PM2.5'].sum().reset_index()
```

```python
china = df[df.Country == 'China'].groupby(['Country','Year'])['PM2.5'].sum().reset_index()
```

```python
plt.plot(iran.Year,iran['PM2.5'],linestyle ='dotted',marker='D',color='green',label= 'Iran')
plt.plot(china.Year,china['PM2.5'],linestyle='dashed',marker='>',color='red',label = 'China')
plt.xlabel('Year')
plt.ylabel('PM2.5')
plt.title('Year wise PM2.5 for Iran vs China')
plt.legend()
```

###`Problem-9` Show Quarter wise Sales data for all products as multi Bar chart.


```python
# code here
q1 = df[(df.month_number >= 1) & (df.month_number <=3)]
q2 = df[(df.month_number >= 4) & (df.month_number <=6)]
q3 = df[(df.month_number >= 7) & (df.month_number <=9)]
q4 = df[(df.month_number >= 10) & (df.month_number <=12)]
```

```python
temp_df = pd.concat([q1.sum() , q2.sum() ,q3.sum() ,q4.sum()],axis=1).reset_index().rename(columns={'index':'Products',0:'q1',1:'q2',2:'q3',3:'q4'})
```

```python
temp_df.drop([0,7,8],inplace=True)
```

```python
temp_df
```

```python
plt.bar(np.arange(temp_df.shape[0])-0.3,temp_df['q1'],width=0.2,color='red')
plt.bar(np.arange(temp_df.shape[0])-0.1,temp_df['q2'],width=0.2,color='blue')
plt.bar(np.arange(temp_df.shape[0])+0.1,temp_df['q3'],width=0.2,color='green')
plt.bar(np.arange(temp_df.shape[0])+0.3,temp_df['q4'],width=0.2,color='brown')
plt.xticks(np.arange(temp_df.shape[0]),temp_df.Products,rotation=45)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Quarterly products sales')
```

###`Problem-10` Plot Stacked Bar chart quarter wise for each product.

```python
# code here
plt.bar(temp_df.Products,temp_df.q1)
plt.bar(temp_df.Products,temp_df.q2,bottom=temp_df.q1)
plt.bar(temp_df.Products,temp_df.q3,bottom=(temp_df.q1+temp_df.q2))
plt.bar(temp_df.Products,temp_df.q4,bottom=(temp_df.q1+temp_df.q2+temp_df.q3))
plt.xticks(rotation=45)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Quarterly products sales')
```
