```python
#function to convert playtime in hours
def timecorrection(a):
    if a.startswith('Less'):
        hrs = round(1/60,3)
    else:
        hrs = int(a.split()[0])
        try:
            min = int(a.split()[3])
        except:
            min = 0
        hrs = hrs + round(min/60,3)
    return hrs
s='2 hrs and 20 mins'
timecorrection(s)
```

```python
_.time = _.time.apply(lambda x: timecorrection(x))
```
