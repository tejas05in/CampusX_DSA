### **Problem-1:** Make a subplots which have 2 plots.

- For the first chart, draw a scatter plot "*Monitored Cap.(MW)*" vs "*Total Cap. Under Maintenance (MW)*" of top 5 most frequent power stations. Then draw the lines which indicate the average values of these two columns. Change the colors according to the names of the Power Stations.
- For the second chart, draw a scatter plot "*Monitored Cap. (MW)*" vs "*Actual(MU)*" of the top 5 most frequent power stations. Also draw the lines which indicates the average values of these two columns. Change the colors according to the names of the Power Stations.


````python
# code here
df = pd.read_csv("PowerGeneration - PowerGeneration.csv")
````

````python
df.head()
l = list(df['Power Station'].value_counts()[:5].index)
temp = df[df['Power Station'].isin(l)]
````

````python
temp = temp.reset_index(drop=True)
````

````python
d={}
count = 0
for i in l:
    d[i] = count
    count+=1
print(d)
````

````python
temp['Power Station']= temp['Power Station'].replace(d)
````

````python
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(15,9))

ax[0].scatter(temp['Monitored Cap.(MW)'],temp['Total Cap. Under Maintenace (MW)'],c=temp['Power Station'])
ax[0].axvline(temp['Monitored Cap.(MW)'].mean(),c='red')
ax[0].axhline(temp['Total Cap. Under Maintenace (MW)'].mean(),c='red')
ax[0].set_xlabel("Monitored capacity in Mega Wats")
ax[0].set_ylabel("Total Capacity under maintainence in Mega Wats")
ax[0].set_title("Monitored Capacity vs Total Capacity under maintainance for top 5 power stations in MW")

ax[1].scatter(temp['Monitored Cap.(MW)'],temp['Actual(MU)'],c=temp['Power Station'])
ax[1].axvline(temp['Monitored Cap.(MW)'].mean(),c='red')
ax[1].axhline(temp['Actual(MU)'].mean(),c='red')
ax[1].set_xlabel("Monitored capacity in Mega Wats")
ax[1].set_ylabel("Actual in Mega Wats")
ax[1].set_title("Monitored Capacity vs Actual Capacity for top 5 power stations in MW")
````

### **Problem-3:** Make a 3D *Surface* plot of this below mathematical equation.

$$z = |x| - |y|$$

````python
# code here
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
````

````python
xx , yy = np.meshgrid(x,y)
````

````python
z = abs(xx) - abs(yy)
````

````python
z.shape
````

````python
fig = plt.figure()
ax = plt.subplot(projection='3d')
p = ax.plot_surface(x,y,z,cmap='plasma')
fig.colorbar(p)
````

### **Problem-4:** Draw the 3D *Contour plot* of this below equation:

$$z = |x| - |y|$$

````python
# code here
fig = plt.figure()
ax = plt.subplot()
p = ax.contour(x,y,z,cmap='plasma')
fig.colorbar(p)
````
