```python
import seaborn as sns
```

```python
# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
#axis level function
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
```

```python
#fig level function
sns.catplot(tips,x='day',y='total_bill',kind='strip')
```

```python
sns.catplot(tips,x='day',y='total_bill',kind='strip',jitter=0.2,hue='sex')
```

```python
#swarmplot
sns.catplot(tips,x='day',y='total_bill',kind='swarm',hue='sex')
```

```python
sns.swarmplot(tips,x='day',y='total_bill',hue='sex')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box',hue='sex')
```

```python
sns.boxplot(data=tips,y='total_bill')
```

```python
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

```python
sns.barplot(data=tips,x='sex',y='total_bill',errorbar=None)
```

```python
import numpy as np
sns.catplot(data=tips,x='sex',y='total_bill',errorbar=None,kind='bar',hue='smoker',estimator=np.min)
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='point',hue='smoker',errorbar=None)
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='bar',errorbar=None)
```

```python
sns.countplot(tips,x='sex',hue='day')
```

```python
sns.catplot(tips,x='sex',y='total_bill',col='smoker',kind='box',row='time')
```

```python
#regplot
sns.regplot(tips,y='tip',x='total_bill')
```

```python
sns.lmplot(tips,y='tip',x='total_bill',hue='sex')
```

```python
#residpolt
sns.residplot(tips,x='total_bill',y='tip')
```

```python
# figure level -> relplot,displot,catplot,lmplot
sns.catplot(tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

```python
g= sns.FacetGrid(tips,col='day',row='time',hue='smoker')
g.map(sns.violinplot,'sex','total_bill')
g.add_legend()
```

```python
sns.pairplot(iris,hue='species')
```

```python
g = sns.PairGrid(iris,hue='species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
g = sns.PairGrid(iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
g = sns.PairGrid(iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
sns.jointplot(tips,x='total_bill',y='tip',kind='hist',hue='sex')
```

```python
g = sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.kdeplot,sns.violinplot)
```
