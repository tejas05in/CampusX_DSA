```python
import time
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
final = pd.DataFrame()
for i in range(1,42):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.5414.119 Safari/537.36'}
    webpage = requests.get(f'https://www.flipkart.com/search?q=smartphones&page={i}',headers=headers).text
    soup = BeautifulSoup(webpage,'lxml')
    phone = soup.find_all('div',class_='_2kHMtA')
    img_url = []
    img_name = []
    avg_rating = []
    total_rating = []
    total_reviews = []
    disc_price = []
    actual_price = []
    for i in phone:
        img_url.append(i.find('img',class_="_396cs4").get('src'))
        img_name.append(i.find('img',class_="_396cs4").get('alt'))
        try:
            avg_rating.append(float(i.find('div',class_="_3LWZlK").text))
        except:
            avg_rating.append(np.nan)
        try:
            total_rating.append(int(i.find('span',class_="_2_R_DZ").text.split()[0].replace(",","")))
        except:
            total_rating.append(np.nan)
        try:
            total_reviews.append(int(i.find('span',class_="_2_R_DZ").text.split()[3].replace(",","")))
        except:
            total_reviews.append(np.nan)
        try:
            disc_price.append(int(i.find('div',class_="_30jeq3 _1_WHN1").text.strip('').replace(",","")))
        except:
            disc_price.append(np.nan)
        try:
            actual_price.append(int((i.find('div',class_="_3I9_wc _27UcVY").text.strip('').replace(",",""))))
        except:
            actual_price.append(np.nan)
    d = {'image_url':img_url,'image_name':img_name,'average_ratings':avg_rating,'total_ratings':total_rating,'total_reviews':total_reviews,'discounted_price':disc_price,'actual_price':actual_price}
    df = pd.DataFrame(d)
    final  = final.append(df,ignore_index=True)
    time.sleep(15)
    
```
