### 2D Line plot

image.png

- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]


plt.plot(year,price)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#9EFF36')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#D736FF')
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#9EFF36',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#D736FF',linestyle='dashdot')
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#9EFF36',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#D736FF',linestyle='dashdot',linewidth=2)
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#9EFF36',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#D736FF',linestyle='dashdot',linewidth=2,marker='>')
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#9EFF36',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#D736FF',linestyle='dashdot',linewidth=2,marker='>',label='Rohit')
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='upper right')
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]


plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
# grid

plt.plot(batsman['index'],batsman['V Kohli'],color='#9EFF36',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#D736FF',linestyle='dashdot',linewidth=2,marker='>')

plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')


plt.grid()
```

```python
# show
plt.plot(batsman['index'],batsman['V Kohli'],color='#9EFF36',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#D736FF',linestyle='dashdot',linewidth=2,marker='>')

plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')


plt.grid()
plt.show()
```

```python
# plt.scatter on pandas data
df = pd.read_csv('batter.csv')
df=df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='+')
plt.title('Average and Strike Rate analysis of top 50 batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children =[10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']
plt.bar(colors,children,color='black')
```

```python
# bar chart using data
```

```python
# horizontal bar chart
plt.barh(colors,children,color='black')
```

```python
# color and label
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]),df['batsman'])

plt.show()
```

```python
np.arange(df.shape[0])
```

```python
# Multiple Bar charts
```

```python
# xticks
```

```python
# a problem
children =[10,20,40,10,30]
colors = ['red red red red red red red red','blue blue blue blue blue blue','green green green green green green','yellow yellow yellow yellow yellow yellow','pink pink pink pink pink pink ']
plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2016']+df['2015']),label='2017')
plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects=['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)
plt.show()
```

```python
# dataset
df = pd.read_csv('gayle-175.csv')
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# percentage and colors
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'],explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
